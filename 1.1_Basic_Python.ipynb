### Задание 1
<p>
Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
</p>
<p>
Напишите код, который проверяет какая из этих строк длиннее.
</p>

```python
phrase_1 = str(input('Введите первую фразу: '))
phrase_2 = str(input('Введите вторую фразу: '))

if len(phrase_1) > len(phrase_2):
    print('Фраза 1 длиннее фразы 2')

elif len(phrase_1) < len(phrase_2):
  print('Фраза 2 длиннее фразы 1')
  
else:
  print('Фразы равной длины')
```

### Задание 2
<p>
Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.
</p>

```python
# год является високосным, если его номер кратен 4, но не кратен 100, а также если он кратен 400

year = int(input('Введите год '))

if year <= 1000 or year >= 9999: #ограничение по кол-ву символов для ввода
  print('Неверные данные')

elif (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Високосный")

else:
    print("Обычный")
```

### Задание 5 (необязательное)
<p>
Дана переменная, в которой хранится шестизначное число (номер проездного билета). 
</p>
<p>
Напишите программу, которая будет определять, является ли данный билет “счастливым”. </p>
<small>
Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.</small>

```python
print('Введите номер Вашего билета:')

n = input()

sum_left = int(n[0]) + int(n[1]) + int(n[2]) #сумма первых трех цифр
sum_right = int(n[3]) + int(n[4]) + int(n[5]) #сумма последних трех цифр

if sum_left == sum_right:
  print('Счастливый')

else:
  print('Обычный') 
  
```

### Задание 6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

```python
#вариант 1
user_input = input('Введите название фигуры: ')
figures = ['треугольник', 'круг', 'прямоугольник']

if user_input in figures:
  if user_input == 'круг':
    rad = input('Введите радиус: ')
    sq_circle = 3.14 * (rad) ** 2
    print(sq_circle)

  elif user_input == 'треугольник':
    a = [float(x) for x in input('Введите длины сторон треугольника: ').split(',')]
    p = (a[0] + a[1] + a[2]) / 2
    sq_triang = (p * (p - a[0]) * (p - a[1]) * (p - a[2])) ** 0.5
    print(sq_triang)

  elif user_input == 'прямоугольник':
    rect = [float(x) for x in input('Введите длины сторон прямоугольника: ').split(',')]
    sq_rect = rect[0] * rect[1]
    print(sq_rect)

  else:
    print('Неверный ввод')
```
