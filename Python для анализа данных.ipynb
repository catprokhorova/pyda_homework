```python
# год является високосным, если его номер кратен 4, но не кратен 100, а также если он кратен 400

year = int(input('Введите год '))

if year <= 1000 or year >= 9999: #ограничение по кол-ву символов для ввода
  print('Неверные данные')

elif (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Високосный")

else:
    print("Обычный")
```

```python
queries = [
  'смотреть сериалы онлайн',
  'новости спорта',
  'афиша кино',
  'курс доллара',
  'сериалы этим летом',
  'курс по питону',
  'сериалы про спорт',
]

storage = {} #create empty dict

for query in queries:
    words = query.split() #split every query by " "

    if len(words) in storage.keys():
        storage[len(words)] += 1
    else:
        storage.update({len(words): 1}) #fill the storage

for key, value in storage.items():
    result = round((value / len(queries)) * 100, 2)
    print(f'Поисковых запросов из {key} слова: {result} % ')
```

###Задание 1
<p>Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру.</p>
<p>Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится.</p>
<p>Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам.</p>
<p>Пункт 4. Пользователь по команде “ads” может добавить новую полку.</p>
<p>Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая).</p>



```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
def find_owner(): #command "p"
  number = input('Введите номер документа: ')
  for i in documents:
    if i['number'] == number:
      return 'Владелец документа: ' + i['name']

  return 'Документ не найден в базе'
```

```python
def shelf(number): #command "s"
  if number == '':
    number = input('Введите номер документа: ')
  for i in directories:
    if number in directories[i]:
      return f'Документ хранится на полке: {i}'

  return 'Документ не найден в базе'
```

```python
def info(): #command "l"
  all = ''
  for doc in documents:
    all += f'№: {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {shelf(doc["number"])}' + '\n'
  return all
```

```python
def add_shelf(number): #command "ads"
  if number not in directories:
    directories[number] = []
    return 'Полка добавлена. Текущий перечень полок:' + ','.join(directories.keys())
  return 'Такая полка уже существует. Текущий перечень полок: ' + ','.join(directories.keys())
```

```python
def del_shelf(): #command "ds"
  number = input('Введите номер полки: ')
  if number in directories:
    if len(directories[number]) > 0:
      return 'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + ', '.join(directories.keys())
    else:
      del directories[number]
      return 'Полка удалена. Текущий перечень полок: ' + ', '.join(directories.keys())
  else:
    return 'Такой полки не существует. Текущий перечень полок:' + ', '.join(directories.keys())
```

```python
def main():
  user_input = input('Введите команду: ')
  
  while user_input != 'q':
    if user_input == 'p':
      print(find_owner())
    elif user_input == 's':
      print(shelf(""))
    elif user_input == 'l':
      print(info())
    elif user_input == 'ads':
      print(add_shelf())
    elif user_input == 'ds':
      print(del_shelf())
    else:
      print('Неверная команда')
  
    print(documents)
    print(directories)
            
main()
```

###Задание 2
Добавьте в класс Rate параметр diff (со значениями True или False), который в случае значения True в методах курсов валют (eur, usd итд) будет возвращать не курс валюты, а изменение по сравнению в прошлым значением. Считайте, self.diff будет принимать значение True только при возврате значения курса. 

При отображении всей информации о валюте он не используется.

```python
import requests
class Rate: #copied entire class from lecture notebook
    def __init__(self, format_='value', diff = False):
        self.format = format_
        self.diff = diff
    
    def exchange_rates(self):
        self.r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r.json()['Valute']
    
    def make_format(self, currency):
        response = self.exchange_rates()
        
        if currency in response:
            if self.format == 'full':
              return response[currency]

            if self.format == 'value' and self.diff is True:
              return response[currency]['Value'] - response[currency]['Previous']
            
            if self.format == 'value':
              return response[currency]['Value']
        
        return 'Error'
    
    def eur(self):
        return self.make_format('EUR')
    
    def usd(self):
        return self.make_format('USD')
    
    def brl(self):
        return self.make_format('BRL')
```

```python
r = Rate(format_= 'full')
print(r.usd())
print('\n***********\n')

r = Rate(diff = True)
print(r.usd())
print('\n***********\n')

r = Rate(diff = False)
print(r.usd())
```

###Задание 4
Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

<i>Пример работы программы:

emails = [‘test@gmail.com’, ‘xyz@test.in’, ‘test@ya.ru’, ‘xyz@mail.ru’, ‘xyz@ya.ru’, ‘xyz@gmail.com’]

Результат:

- gmail.com: 2
- test.in: 1
- ya.ru: 2
- mail.ru: 1</i>

```python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']

def counting(emails):
  res = {}
  pattern = re.findall(r'(?<=@)(\w+\.\w+)', str(emails))
  for i in pattern:
    res.setdefault(i, 0)
    res[i] += 1
  
  return res


counting(emails)
```
